**How to clean Data:**

To clean the data we have to follow the below steps
- Data Selection
- Handling missing values
- Datatype conversions
- Removing Duplicates

**Data Selection:**
We have two simple funtions to select the data from the table

i)  df.loc["Row name","column name"] ---> This function is used when you want to select data based on row name and column name.

ii) df.iloc["Row index","column index"] ---> This function is used when you want to select data based on row index and column index.

```python
#Based on names
df.loc[0:3 , "Loan_ID":"Married"]
```

```python
#Based on index
df.iloc[0:3 , 0:3]  # This will give us the first 3 rows and first 3 columns of the dataframe.
                    #  The index 0 means the first row or column and the index 3 means (0,1,2)---> it does not include the index 3 it takes as 3-1=2. So, it will give us the rows with index 0,1,2 and columns with index 0,1,2.
```

**Handling Missing Values:**
There are three simple steps to be followed 
 
 #Checking Null values
      i) df.isnull().sum() --> Gives all the null values count for each column.

 #Droping null values
      i) df.dropna()---> Drops all null values

 #Filling null values
      i) df.fillna("value to be filled") --> fills the null value with another value

```python
#checking null values
df.isnull().sum()
```

```python
#droping null values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

**Datatype conversions:**
